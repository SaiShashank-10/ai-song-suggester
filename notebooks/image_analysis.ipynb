```python
# Import the necessary libraries
import tensorflow as tf
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50 import preprocess_input
import numpy as np
from PIL import Image
import os

# Load the pre-trained ResNet50 model
# 'include_top=False' prevents loading the final classification layer.
# 'pooling='avg'' averages the features into a single vector.
print("Loading the pre-trained ResNet50 model...")
model = ResNet50(weights='imagenet', include_top=False, pooling='avg')
print("✅ Model loaded successfully!")

def extract_features(img_path):
    """
    Takes an image path, loads the image, and extracts a feature vector using ResNet50.
    """
    # Load the image and resize it to 224x224 pixels, as required by ResNet50
    img = Image.open(img_path).resize((224, 224))
    
    # Convert the image to a 3-channel image (RGB) if it has a different mode (like RGBA or grayscale)
    if img.mode != "RGB":
        img = img.convert("RGB")

    # Convert the image into a numpy array (a grid of numbers)
    x = np.array(img)
    
    # Add an extra dimension to the array to make it a "batch" of one image
    x = np.expand_dims(x, axis=0)
    
    # Pre-process the image for the ResNet50 model (this normalizes the pixel values)
    x = preprocess_input(x)
    
    # Get the feature vector from the model
    features = model.predict(x)
    
    # Flatten the features into a simple 1D array
    return features.flatten()

# --- HOW TO USE THE FUNCTION ---

# 1. Find an image on your computer (.jpg, .png, etc.).
# 2. Place it inside your 'notebooks' folder.
# 3. Replace 'your_image.jpg' below with the actual name of your image file.
image_path = 'image.png'  # Update this path

# Safety check to make sure the file exists before processing
if os.path.exists(image_path):
    # Call the function to get the image's "fingerprint"
    features = extract_features(image_path)
    
    # Print the results
    print(f"\nSuccessfully extracted features for '{image_path}'")
    print(f"Feature vector shape: {features.shape}")
    print("\nHere are the first 10 features as an example:")
    print(features[:10])
else:
    print(f"\n❌ ERROR: The file '{image_path}' was not found in the 'notebooks' folder.")
    print("Please make sure you have placed an image there and updated the 'image_path' variable.")
```
